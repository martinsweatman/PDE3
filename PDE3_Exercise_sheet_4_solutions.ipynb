__Q1__
(Adapted from exercises 9.5.2 of Advanced Modern Engineering Mathematics)

a) Show that the function $\phi(x,y)=x^2y$ satisfies the 2-d Poisson equation

\begin{align}
\nabla^2 \phi = \frac{\partial^2 \phi(x,y)}{\partial x^2} + \frac{\partial^2 \phi(x,y)}{\partial y^2} = 2y & & (1)
\end{align}

__Solution__

First, find the first and second partial differentials of $\phi(x,y)$

\begin{align}
\frac{\partial \phi(x,y)}{\partial x} &= 2xy & \frac{\partial^2 \phi(x,y)}{\partial x^2} &= 2y & \\
\frac{\partial \phi(x,y)}{\partial y} &= x^2 & \frac{\partial^2 \phi(x,y)}{\partial y^2} &= 0 & 
\end{align}

Then substitute these into (1) to find

\begin{align}
\frac{\partial^2 \phi(x,y)}{\partial x^2} + \frac{\partial^2 \phi(x,y)}{\partial y^2} = 2y
\end{align}

thus showing the function $\phi(x,y)=x^2y$ satisfies the 2-d Poission equation (1).

b) Now let $\phi(x,y) = u(x,y) + x^2y$, where $u(x,y)$ satisfies the Laplace equation $\nabla^2 u = 0$. Show that $\phi(x,y)$ still satisfies the Poisson equation (1).
\
\
__Solution__

If $\nabla^2 u = 0$, then

$$
\nabla^2 \phi(x,y) = \nabla^2 u(x,y) + \nabla^2 (x^2 y) = 0 + 2y = 2y
$$

Thus, if $u(x,y)$ satisfies the Laplace equation then $\phi (x,y)=u(x,y)+x^2y$ must satisy the Poisson equation (1).
\
\
c) Use the method of separated solutions to find the solution for $\phi(x,y)$ in the unit square which satisfies the boundary conditions

\begin{align}
\phi(0, y) &= 0, & 0 \le y \le 1 && (2) \\
\phi(1, y) &= y, & 0 \le y \le 1 && (3) \\
\phi(x, 0) &= 0, & 0 \le x \le 1 && (4) \\
\phi(x, 1) &= x^2 + \sin\pi x, & 0 \le x \le 1 && (5)
\end{align}
\
\
__Solution__

First note that the source term $x^2y$ from part a) is generated by integration of the inhomogeneity $2y$ twice with respect to $x$ and setting the integration constants to $0$. Other more general forms of the source term could have been achieved by integration of the inhomogeneity with respect to $x$ and $y$. However, the specific source term $x^2y$ is consistent with the boundary conditions (2)-(5) except for the trigonemtric term in $x$ in BC (5). Since this trigonometric term cannot be accounted for by choosing other values for the integration constants, but it can be generated by solving the Laplace equation for $u(x,y)$ (see below), we can be satisfied that the specific source term $x^2y$ is the only source term that satisfies these boundary conditions.

To solve the Laplace equation for $u(x,y)$ we must first transform the boundary conditions (2)-(5)

\begin{align}
u(0, y) &= \phi(0,y) - 0 = 0-0 = 0, & 0 \le y \le 1 && (6) \\
u(1, y) &= \phi(1,y) - y = y-y = 0, & 0 \le y \le 1 && (7) \\
u(x, 0) &= \phi(x,0) - 0 = 0-0 = 0, & 0 \le x \le 1 && (8) \\
u(x, 1) &= \phi(x,1) - x^2 = x^2 + \sin\pi x - x^2 = \sin\pi x, & 0 \le x \le 1 && (9)
\end{align}

Now we can solve the Laplace equation for $u(x,y)$. From the lectures or the book we know that the following three functions

\begin{align}
u_1(x,y) =& (A\sin \mu x+B\cos \mu x) \left(Ce^{\mu y}+De^{-\mu y}\right) \\
u_2(x,y) =&\left(Ae^{\mu x}+Be^{-\mu x}\right)(C\sin \mu y+D\cos \mu y) \\
u_3(x,y) =& (Ax+B)(Cy+D)
\end{align}

are possible solutions for the Laplace equation.

We need to go through the boundary conditions to find which of these three functions is a solution to the partial differential equation with associated boundary conditions. In addition, we need to use the boundary conditions to find the values for the parameters, i.e. $A$, $B$, $C$, $D$ and $\mu$, in these functions.

The first two boundary conditions (6) and (7) require that the solution has $\sin$ and $\cos$ terms in $x$ since these terms have multiple zeros. Conversely, $\sinh$ and $\cosh$ have only $1$ or $0$ roots and are thus not suitable for the term in $x$ with these boundary conditions. Also $u_3$ is not suitable because it can only give zeros at $x=0$ and $x=1$ for all $y$ if $u_3(x,y)=0$, which contradicts boundary condition (9). Thus the solution must have the form,

$$
u_1(x,y) = (A\sin \mu x+B\cos \mu x) \left(Ce^{\mu y}+De^{-\mu y}\right)
$$


Note that if the boundary conditions (2)-(5) had additional linear terms in $x$ or $y$ they could have been accounted for either by considering a more general source term or by including $u_3(x,y)$ in the solution for $u(x,y)$. Either approach would lead to the same unique solution that satisies the boundary conditions.

Dropping the subscript and using boundary condition (6) we get

\begin{align}
u(0,y) &= (A\sin(\mu 0)+B\cos(\mu 0)) \left(Ce^{\mu y}+De^{-\mu y}\right) \\
 &= B \left(Ce^{\mu y}+De^{-\mu y}\right) \overset{!}{=} 0 
\end{align}

This needs to hold for all $0 \leq y \leq 1$ which is only possible for $B=0$. We can thus write the solution as

$$
u(x,y) = \sin(\mu x) \left(\tilde{C}e^{\mu y} + \tilde{D}e^{-\mu y}\right)
$$

where $\tilde{C} = AC$ and $\tilde{D} = AD$.

Using boundary condition (7) we get,

$$
u(1,y) = \sin(\mu) \left(\tilde{C}e^{\mu y} + \tilde{D}e^{-\mu y}\right) \overset{!}{=} 0 
$$

Again this needs to hold for all $0 \leq y \leq 1$ which can only be fulfilled if $\sin(\mu)=0$. We know that the sine function is zero for multiples of $\pi$ and thus that $\mu=n\pi$ where $n=1,2,\dots$.

Using boundary condition (9) we get

$$
u(x,1) = \sum_{n=0}^{\infty} \sin(n\pi x) \left(\tilde{C_n}e^{n\pi} + \tilde{D_n}e^{-n\pi}\right) \overset{!}{=} \sin{\pi x} 
$$

which holds for $0\le x \le 1$ only if we have $n=1$ and $\left(\tilde{C_1}e^{\pi} + \tilde{D_1}e^{-\pi}\right) =1$. Thus the solution is

\begin{align}
u(x,y) = \sin(\pi x) \left(\tilde{C_1}e^{\pi y} + \tilde{D_1}e^{-\pi y}\right)
\end{align}

To satisfy boundary condition (8)

$$
u(x,0) = \sin(\pi x) \left(\tilde{C_1}e^{\pi 0} + \tilde{D_1}e^{-\pi 0}\right) \overset{!}{=} 0, \quad 0\le x \le1
$$

we must have $\left(\tilde{C_1}e^{\pi 0} + \tilde{D_1}e^{-\pi 0}\right) = 0$, which means $\tilde{C_1}=-\tilde{D_1}$. Therfore,
using the above constraint $\left(\tilde{C_1}e^{\pi} - \tilde{C_1}e^{-\pi}\right) =1$, which means $\tilde{C_1} = 0.5(\sinh{\pi})^{-1}$. The full solution is then,

$$
\phi(x,y) = \frac{\sin{\pi x} \sinh{\pi y}}{\sinh{\pi}} + x^2y
$$


d) Use Python to plot the solution.

__Solution__

The Python script is the same as for Exercise sheet 1 with the exception of a different function definition and different solution domain.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator , FormatStrFormatter
import numpy as np
# Solution to the Poisson equation on the unit square
# Set the x and y grid points and the meshgrid
N = 100
x = np.linspace(0,1,N)
y = np.linspace(0,1,N)
X,Y = np.meshgrid(x,y)
# Define the function and generate the output on the meshgrid
Z = np.sin(np.pi * X) * np.sinh(np.pi * Y) / np.sinh(np.pi) + X*X*Y
# Generate the plot
fig = plt.figure()
ax = plt.axes(projection='3d')
# Plot the surface
surf = ax.plot_surface(X,Y,Z,cmap=cm.coolwarm,linewidth=0,antialiased=False)
# Customize the z axis
ax.set_zlim(0,2)
ax.zaxis.set_major_locator(LinearLocator(6))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.01f'))
ax.set_xlabel('xaxis')
ax.set_ylabel('yaxis')
ax.set_zlabel('phi(x,y)')
# Add a color bar which maps values to colors
fig.colorbar(surf,shrink=0.75,aspect=8,location="left")
plt.show()
```

__Q2__
This question concerns steady-state heat flow radially within a long cyclindrical pipe with uniform material properties.
Heat flow is modelled with the Laplace equation on an annulus with inner and outer radii $r=1$ and $r=2$
(all parameters are in non-dimensional form).

The boundary conditions on the non-dimensional temperature profile, $T(r,\theta)$, are,

\begin{align}
T(1,\theta) &= 2\sin^2(\theta) &= 0\le\theta<\pi && (6) \\
T(2,\theta) &= 0 &= 0\le\theta<\pi && (7)
\end{align}

a) Derive the analytical solution of the Laplace equation for the nondimensional temperature on the annulus with the given boundary conditions using the method of separated solutions. Note, the Lapalace equation in cyclindrical coordinates is given by,

\begin{align}
\frac{\partial^2 T}{\partial r^2} + \frac{1}{r}\frac{\partial T}{\partial r} + \frac{1}{r^2}\frac{\partial^2 T}{\partial \theta^2} = 0 && \quad (8)
\end{align}

with a separated solution of the form $T(r,\theta) = R(r) \Theta(\theta)$.
\
\
__Solution__

From the course notes we know there are three general solutions to the Laplace equation in cylindrical coordinates. Let's consider them in turn.

First, for $\lambda=0$ we have the combined solution,

$$
T_0(r, \theta) = (a\theta + b)(C\ln(r) + D)
$$

We know that the solution needs to be periodic in $2\pi$ and this is only the case for $a=0$.

Thus we are left with,

$$
T_0(r,\theta) = k \ln(r/d)
$$

where $k=bC$ and $d$ are constants which need to be determined. 

For $\lambda<0$ we get the combined solution,

\begin{align}
T_-(r,\theta) &= \left(A\cosh(\mu\theta) + B\sinh(\mu\theta)\right) \left(C r^{-\mu} + D r^{\mu}\right)
\end{align}

with parameters $A$, $B$, $C$ and $D$, and where $\lambda = -\mu^2$. Again, this needs to be periodic in $2\pi$ which is only fulfilled for $A=B=0$. Thus the solution for $\lambda<0$ is zero. 

For $\lambda = \mu^2 > 0$ we get the combined solution,

\begin{align}
T_+(r,\theta) &= \left(A\cos(\mu\theta) + B\sin(\mu\theta)\right) \left(C r^{-\mu} + D r^{\mu}\right)
\end{align}

with different parameters $A$, $B$, $C$ and $D$. Again, this needs to be periodic in $2\pi$ and this is the case only for $\mu=n$ with $n=0,1,2,...$

Thus the solution is,

$$
T_+(r, \theta) = \left(A_n\cos(n\theta) + B_n\sin(n\theta)\right) \left(C_n r^{-n} + D_n r^{n} \right) 
$$

where the constants $A_n$, $B_n$, $C_n$ and $D_n$ are to be determined. 

To fulfill the boundary conditions we need to superimpose the solutions for $\lambda=0$ and $\lambda>0$. This is possible because we are dealing with a linear PDE with constant coefficients. The complete solution is thus,

$$
T(r, \theta) = k\log(r/d) + \sum_{n=1}^\infty \left(A_n\cos(n\theta) + B_n\sin(n\theta)\right) \left( C_n r^{-n} + D_n r^{n} \right) 
$$

where the term with $n=0$ is omitted because it is redundant.

To satisy the outer boundary condition we have,

$$
T(2, \theta) = k\log(2/d) + \sum_{n=1}^\infty \left(A_n\cos(n\theta) + B_n\sin(n\theta)\right) \left( C_n 2^{-n} + D_n 2^{n} \right) \overset{!}{=}0
$$

Here, we have two options. Either the term in brackets, $\left( C_n 2^{-n} + D_n 2^{n} \right)$, is always zero or it is in general non-zero. If it is generally non-zero then we effectively have the Fourier series with period $2\pi$,

$$
T(2, \theta) = E + \sum_{n=1}^\infty \left(F_n\cos(n\theta) + G_n\sin(n\theta)\right)  \overset{!}{=}0
$$

We know in this case that all the coefficients, $E$, $F_n$ and $G_n$ are zero, so this is not a useful solution. Therefore, we must choose the other option where the term in brackets is always zero. Thus, we require $d=2$ and $C_n 2^{-n} = - D_n 2^{n}$. Therefore the solution simplifies to

$$
T(r, \theta) = k\log(r/2) + \sum_{n=1}^\infty \left(\tilde{A_n} \cos(n\theta) + \tilde{B_n} \sin(n\theta)\right) \left( r^{-n} - r^{n} 2^{-2n} \right)
$$

To satisy the inner boundary condition we then have,

\begin{align}
T(1,\theta) &= k\log(1/2) + \sum_{n=1}^\infty \left(\tilde{A_n}\cos(n\theta) + \tilde{B_n}\sin(n\theta)\right) \left(1 - 2^{-2n}\right) \overset{!}{=} 2\sin^2(\theta) = 1 - \cos(2\theta)
\end{align}

The parameters $k$, $\tilde{A_n}$ and $\tilde{B_n}$ can be found directly. Thus, $k=(\log(1/2))^{-1}$, $\tilde{B_n}=0$, $\tilde{A}_{n \neq 2}=0$, and $\tilde{A_2}(1 - 2^{-4}) = -1$. Therfore, $\tilde{A_2} = -16/15$.

Thus the complete solution is,

\begin{align}
T(r,\theta) &= \frac{\log(r/2)}{\log(1/2)} -  \frac{16}{15} \cos(2\theta) \left(\frac{1}{r^{2}} - \frac{r^{2}}{16} \right) 
\end{align}

b) Use Python to plot the solution.

__Solution__
The Python script is the same as for Exercise sheet 1 with the exception that we calculate the solution in cylindrical coordinates which are then mapped to cartesian coordinates for plotting.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator , FormatStrFormatter
import numpy as np
# Solution to the heat conduction equation in a cyclindrical annulus
# Set the r and theta grid points and the meshgrid
N = 100
a = 1
b = 2
r = np.linspace(a,b,N)
t = np.linspace(0,2*np.pi,N)
R,T = np.meshgrid(r,t)

# Define the function and generate the output on the meshgrid
Z = np.log(R/2)/np.log(0.5) - 16*np.cos(2*T)*(1/(R*R) - (R*R)/16)/15

# Generate the plot
fig = plt.figure()
ax = plt.axes(projection='3d')
# Convert to cartesian
X = R * np.cos(T)
Y = R * np.sin(T)

# Plot the surface
surf = ax.plot_surface(X,Y,Z,cmap=cm.coolwarm,linewidth=0,antialiased=False)
# Customize the z axis
ax.set_zlim(0,2)
ax.zaxis.set_major_locator(LinearLocator(6))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.01f'))
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('T(x,y)')
ax.view_init(35,-15)
# Add a color bar which maps values to colors
fig.colorbar(surf,shrink=0.75,aspect=8, location="left")
plt.show()
```
